```python
import pandas as pd

jobs = pd.read_csv('700-jobs-data-of-ai-and-data-fields-2025/jobs_dataset.csv')
print("Jobs shape:", jobs.shape)
print("Jobs columns:", jobs.columns.tolist())
jobs.head(5)
```

```python
import re

# Rename columns to match our pipeline
resumes = resumes.rename(columns={
    'ID': 'resume_id',
    'Resume_str': 'resume_str',
    'Resume_html': 'resume_html',
    'Category': 'category'
})

# Drop rows missing the actual resume text
resumes = resumes.dropna(subset=['resume_str'])
resumes = resumes.drop_duplicates(subset=['resume_id'])

# Light text cleaning — strip HTML tags and extra spaces
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = re.sub(r'<[^>]+>', ' ', text)       # remove HTML tags
    text = re.sub(r'\s+', ' ', text).strip()   # collapse whitespace
    return text.lower()

resumes['clean_text'] = resumes['resume_str'].apply(clean_text)
resumes['category_norm'] = resumes['category'].str.strip().str.lower()

print("Cleaned resumes:", resumes.shape)
resumes[['resume_id', 'category', 'clean_text']].head(5)
```

```python
# Rename to match our pipeline — using YOUR actual column names
jobs = jobs.rename(columns={
    'positionName':  'position_name',
    'company':       'company',
    'location':      'location',
    'description':   'job_description',
    'salary':        'salary',
    'jobType/0':     'job_type',
    'searchInput/position': 'job_family'
})

# Generate a unique job ID
jobs['job_id'] = ['J' + str(i+1) for i in range(len(jobs))]

# Keep only the columns we need
jobs = jobs[['job_id', 'position_name', 'company', 'location',
             'job_type', 'job_description', 'salary', 'job_family']]

# Drop rows with no job description
jobs = jobs.dropna(subset=['job_description'])

# Clean text
jobs['clean_text'] = jobs['job_description'].apply(clean_text)
jobs['job_family'] = jobs['job_family'].str.strip().str.lower().fillna('general')

print("Cleaned jobs:", jobs.shape)
jobs[['job_id', 'position_name', 'job_family', 'clean_text']].head(5)
```

```python
!pip install pyspark -q

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ATS-Batch-Processing") \
    .getOrCreate()

spark.sparkContext.setLogLevel("WARN")
print("Spark version:", spark.version)
print("Spark is ready!")
```

```python
from pyspark.sql import functions as F

# Load the JSON lines files we saved earlier
raw_resumes = spark.read.json('data/json/resumes.json')
raw_jobs    = spark.read.json('data/json/jobs.json')

print("Raw resumes:", raw_resumes.count())
print("Raw jobs:   ", raw_jobs.count())

raw_resumes.printSchema()
raw_jobs.printSchema()
```

```python
from pyspark.ml.feature import Tokenizer, StopWordsRemover

# Clean resume text
resumes_clean = raw_resumes \
    .withColumn("clean_text", F.lower(F.col("clean_text"))) \
    .withColumn("clean_text", F.regexp_replace("clean_text", "[^a-z\\s]", " ")) \
    .withColumn("clean_text", F.regexp_replace("clean_text", "\\s+", " ")) \
    .withColumn("clean_text", F.trim(F.col("clean_text"))) \
    .dropna(subset=["clean_text"]) \
    .filter(F.col("clean_text") != "")

# Tokenize — split text into individual words
tokenizer = Tokenizer(inputCol="clean_text", outputCol="tokens_raw")
resumes_tokenized = tokenizer.transform(resumes_clean)

# Remove stopwords (the, and, is, etc.)
remover = StopWordsRemover(inputCol="tokens_raw", outputCol="tokens")
resumes_processed = remover.transform(resumes_tokenized)

# Count tokens per resume
resumes_processed = resumes_processed.withColumn(
    "token_count", F.size(F.col("tokens"))
)

print("Processed resumes:", resumes_processed.count())
resumes_processed.select("resume_id", "category", "token_count").show(5)
```

```python
# Clean job text
jobs_clean = raw_jobs \
    .withColumn("clean_text", F.lower(F.col("clean_text"))) \
    .withColumn("clean_text", F.regexp_replace("clean_text", "[^a-z\\s]", " ")) \
    .withColumn("clean_text", F.regexp_replace("clean_text", "\\s+", " ")) \
    .withColumn("clean_text", F.trim(F.col("clean_text"))) \
    .dropna(subset=["clean_text"]) \
    .filter(F.col("clean_text") != "")

# Tokenize
tokenizer_jobs = Tokenizer(inputCol="clean_text", outputCol="tokens_raw")
jobs_tokenized = tokenizer_jobs.transform(jobs_clean)

# Remove stopwords
remover_jobs = StopWordsRemover(inputCol="tokens_raw", outputCol="tokens")
jobs_processed = remover_jobs.transform(jobs_tokenized)

# Count tokens per job
jobs_processed = jobs_processed.withColumn(
    "token_count", F.size(F.col("tokens"))
)

print("Processed jobs:", jobs_processed.count())
jobs_processed.select("job_id", "position_name", "job_family", "token_count").show(5)
```

# **Build the TF-IDF pipeline**

```python
from pyspark.ml.feature import HashingTF, IDF
from pyspark.ml import Pipeline

# Combine position/category text with clean text for better matching
resumes_model = resumes_processed.withColumn(
    "model_text",
    F.concat_ws(" ", F.col("clean_text"), F.coalesce(F.col("category_norm"), F.lit("")))
)

jobs_model = jobs_processed.withColumn(
    "model_text",
    F.concat_ws(" ", F.coalesce(F.col("position_name"), F.lit("")),
                     F.col("clean_text"),
                     F.coalesce(F.col("job_family"), F.lit("")))
)

# Tokenize the model_text column
tokenizer_model = Tokenizer(inputCol="model_text", outputCol="model_tokens_raw")
remover_model   = StopWordsRemover(inputCol="model_tokens_raw", outputCol="model_tokens")
hashingTF       = HashingTF(inputCol="model_tokens", outputCol="tf_features", numFeatures=10000)
idf             = IDF(inputCol="tf_features", outputCol="features", minDocFreq=2)

pipeline = Pipeline(stages=[tokenizer_model, remover_model, hashingTF, idf])

# Combine resumes and jobs into one corpus to fit the IDF model together
# Explicitly cast doc_id to string to avoid NumberFormatException during union
resumes_tagged = resumes_model.withColumn("doc_id",   F.col("resume_id").cast("string")) \
                               .withColumn("doc_type", F.lit("resume"))
jobs_tagged    = jobs_model.withColumn("doc_id",   F.col("job_id").cast("string")) \
                            .withColumn("doc_type", F.lit("job"))

corpus = resumes_tagged.select("doc_id", "doc_type", "model_text") \
         .union(jobs_tagged.select("doc_id", "doc_type", "model_text"))

# Fit TF-IDF on the full corpus
model        = pipeline.fit(corpus)
corpus_vec   = model.transform(corpus)

print("TF-IDF model fitted on", corpus_vec.count(), "documents")
```

# **Compute cosine similarity scores**

```python
import math
from pyspark.sql.types import DoubleType
from pyspark.ml.linalg import Vector

# Define cosine similarity UDF
def cosine_similarity(v1, v2):
    if v1 is None or v2 is None:
        return 0.0
    a = v1.toArray()
    b = v2.toArray()
    dot    = float(a.dot(b))
    norm_a = math.sqrt(float(a.dot(a)))
    norm_b = math.sqrt(float(b.dot(b)))
    if norm_a == 0.0 or norm_b == 0.0:
        return 0.0
    return dot / (norm_a * norm_b)

cosine_udf = F.udf(cosine_similarity, DoubleType())

# Split vectors back into resumes and jobs
resume_vec = corpus_vec.filter(F.col("doc_type") == "resume") \
                        .select(F.col("doc_id").cast("string").alias("resume_id"),
                                F.col("features").alias("resume_features"))

job_vec = corpus_vec.filter(F.col("doc_type") == "job") \
                     .select(F.col("doc_id").cast("string").alias("job_id"),
                             F.col("features").alias("job_features"))

# Join vectors back to metadata
resumes_vec_full = resumes_model \
    .withColumn("resume_id_str", F.col("resume_id").cast("string")) \
    .join(resume_vec, F.col("resume_id_str") == resume_vec["resume_id"], how="inner") \
    .drop("resume_id_str")

jobs_vec_full = jobs_model.join(job_vec, on="job_id", how="inner")

# Filter resumes to tech/data-relevant categories only
# This makes sense because all 735 jobs are AI/data science roles
relevant_categories = [
    "information-technology",
    "engineering",
    "finance",
    "consultant",
    "business-development",
    "banking",
    "accountant"
]

resumes_relevant = resumes_vec_full.filter(
    F.col("category_norm").isin(relevant_categories)
)

print("Relevant resumes for matching:", resumes_relevant.count())
print("Jobs to match against:        ", jobs_vec_full.count())

# CrossJoin relevant resumes with all jobs
# relevant_resumes x 735 jobs — manageable size
pairs = resumes_relevant.crossJoin(jobs_vec_full)

# Compute cosine similarity score
pairs = pairs.withColumn(
    "cosine_score",
    cosine_udf(F.col("resume_features"), F.col("job_features"))
)

# Drop zero scores
pairs = pairs.filter(F.col("cosine_score") > 0.0)

print("Total resume-job pairs scored:", pairs.count())
```

# **Rank top 5 job matches per resume**

```python
from pyspark.sql import Window

# Drop the duplicate resume_id column that came from the join
# Keep only the one from resumes side
pairs_clean = pairs.drop(resume_vec["resume_id"])

# Also drop duplicate features columns to keep things clean
pairs_clean = pairs_clean.drop("resume_features", "job_features")

# Verify no more ambiguous columns
print("Columns in pairs_clean:")
print(pairs_clean.columns)

# Rank jobs for each resume by cosine score
window_resume = Window.partitionBy("resume_id").orderBy(F.desc("cosine_score"))

top_matches = pairs_clean.withColumn("rank", F.row_number().over(window_resume)) \
                         .filter(F.col("rank") <= 5) \
                         .select(
                             "resume_id",
                             "category_norm",
                             "job_id",
                             "position_name",
                             "company",
                             "cosine_score",
                             "rank"
                         )

print("Total top matches:", top_matches.count())
top_matches.orderBy("resume_id", "rank").show(15, truncate=False)
```

# **"ATS rejections threshold" simulation**

```python
import matplotlib.pyplot as plt
import numpy as np

print("=== ATS Rejection Simulation ===\n")

# Simulate common ATS thresholds
thresholds = [0.05, 0.10, 0.15, 0.20, 0.25, 0.30]

threshold_results = []
total_resumes = matches_pd["resume_id"].nunique()

for t in thresholds:
    # A resume "passes" if its BEST match score exceeds the threshold
    best_per_resume = matches_pd.groupby("resume_id")["cosine_score"].max()
    passed = (best_per_resume >= t).sum()
    rejected = total_resumes - passed
    pass_rate = round((passed / total_resumes) * 100, 1)
    reject_rate = round((rejected / total_resumes) * 100, 1)
    threshold_results.append({
        "threshold": t,
        "passed":    passed,
        "rejected":  rejected,
        "pass_rate": pass_rate,
        "reject_rate": reject_rate
    })
    print(f"  Threshold {t:.2f} → {passed} pass ({pass_rate}%) | "
          f"{rejected} rejected ({reject_rate}%)")

thresh_df = pd.DataFrame(threshold_results)

# Chart
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("ATS rejection simulation — how many candidates get filtered out?",
             fontsize=13)

axes[0].bar([str(t) for t in thresh_df["threshold"]],
            thresh_df["reject_rate"],
            color="tomato", edgecolor="none")
axes[0].set_title("Rejection rate by score threshold")
axes[0].set_xlabel("Cosine score threshold")
axes[0].set_ylabel("% of resumes rejected")
axes[0].axhline(y=50, color='gray', linestyle='--', alpha=0.5,
                label='50% line')
axes[0].legend()
axes[0].grid(axis='y', alpha=0.3)

axes[1].plot([str(t) for t in thresh_df["threshold"]],
             thresh_df["passed"],
             marker='o', color='steelblue', linewidth=2, markersize=8)
axes[1].fill_between(range(len(thresh_df)), thresh_df["passed"],
                     alpha=0.1, color='steelblue')
axes[1].set_title("Resumes surviving each threshold")
axes[1].set_xlabel("Cosine score threshold")
axes[1].set_ylabel("Number of resumes passing")
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('outputs/charts/ats_rejection_simulation.png', dpi=200,
            bbox_inches='tight')
plt.show()
print("\nChart saved!")
```

# **Keyword Gap Analysis (Why resumes fail?)**
### This shows applicants exactly what words are in job descriptions that their resumes are missing, which is what the real reason ATS rejects them

```python
from collections import Counter

print("=== Keyword Gap Analysis ===")
print("What words appear in job postings but NOT in resumes?\n")

# Get all words from jobs
job_words = []
for text in jobs_processed.select("clean_text").toPandas()["clean_text"].dropna():
    job_words.extend(text.split())

# Get all words from resumes
resume_words = []
for text in resumes_processed.select("clean_text") \
        .toPandas()["clean_text"].dropna():
    resume_words.extend(text.split())

job_freq    = Counter(job_words)
resume_freq = Counter(resume_words)

# Words that appear heavily in jobs but rarely in resumes
gap_words = {}
for word, count in job_freq.items():
    if len(word) > 4:  # skip short words
        resume_count = resume_freq.get(word, 0)
        # ratio: how much more common in jobs vs resumes
        ratio = count / max(resume_count, 1)
        if count > 50:  # only meaningful frequency
            gap_words[word] = {
                "job_count":    count,
                "resume_count": resume_count,
                "ratio":        round(ratio, 2)
            }

gap_df = pd.DataFrame(gap_words).T.reset_index()
gap_df.columns = ["keyword", "job_count", "resume_count", "ratio"]
gap_df = gap_df.sort_values("ratio", ascending=False).head(20)

print("Top 20 keywords candidates are MISSING from their resumes:")
print(gap_df[["keyword", "job_count", "resume_count",
              "ratio"]].to_string(index=False))

# Chart
fig, ax = plt.subplots(figsize=(12, 7))
x = range(len(gap_df))
width = 0.4

bars1 = ax.bar([i - width/2 for i in x], gap_df["job_count"],
               width, label="In job postings", color="tomato")
bars2 = ax.bar([i + width/2 for i in x], gap_df["resume_count"],
               width, label="In resumes", color="steelblue")

ax.set_title("Keyword gap: what job postings demand vs what resumes contain",
             fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(gap_df["keyword"], rotation=45, ha='right')
ax.set_ylabel("Word frequency")
ax.legend()
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('outputs/charts/keyword_gap.png', dpi=200, bbox_inches='tight')
plt.show()
print("Chart saved!")
```

```python
from pyspark.sql.types import (
    StructType, StructField, StringType, LongType, DoubleType
)

# Define schema for incoming resume JSON files
stream_schema = StructType([
    StructField("resume_id",    StringType(),  True),
    StructField("resume_str",   StringType(),  True),
    StructField("category",     StringType(),  True),
    StructField("category_norm",StringType(),  True),
    StructField("clean_text",   StringType(),  True),
    StructField("token_count",  LongType(),    True),
])

# Pre-vectorize jobs using our already-fitted TF-IDF model
# so we can score incoming resumes against them instantly
jobs_static = jobs_model.withColumn(
    "model_text",
    F.concat_ws(" ",
        F.coalesce(F.col("position_name"), F.lit("")),
        F.col("clean_text"),
        F.coalesce(F.col("job_family"), F.lit(""))
    )
)

jobs_static_vec = model.transform(jobs_static) \
                       .select("job_id", "position_name", "company",
                               "job_family", F.col("features").alias("job_features"))

jobs_static_vec.cache()
print("Static job vectors ready:", jobs_static_vec.count())
```

## **Define the micro-batch processing function**

```python
from pymongo import MongoClient
import math

def process_batch(batch_df, batch_id):
    """Called automatically by Spark for each new micro-batch of files."""

    if batch_df.rdd.isEmpty():
        print(f"Batch {batch_id}: no new resumes — skipping")
        return

    print(f"\n--- Batch {batch_id}: processing {batch_df.count()} new resumes ---")

    # Vectorize incoming resumes using the same fitted TF-IDF model
    batch_with_text = batch_df.withColumn(
        "model_text",
        F.concat_ws(" ",
            F.coalesce(F.col("clean_text"),    F.lit("")),
            F.coalesce(F.col("category_norm"), F.lit(""))
        )
    )

    batch_vec = model.transform(batch_with_text) \
                     .select(
                         F.col("resume_id"),
                         F.col("category_norm"),
                         F.col("features").alias("resume_features")
                     )

    # CrossJoin this small batch against all static job vectors
    batch_pairs = batch_vec.crossJoin(jobs_static_vec)

    # Score with cosine similarity
    batch_scored = batch_pairs.withColumn(
        "cosine_score",
        cosine_udf(F.col("resume_features"), F.col("job_features"))
    ).filter(F.col("cosine_score") > 0.0)

    # Keep top 3 matches per resume
    window = Window.partitionBy("resume_id").orderBy(F.desc("cosine_score"))
    top3 = batch_scored.withColumn("rank", F.row_number().over(window)) \
                       .filter(F.col("rank") <= 3) \
                       .select("resume_id", "category_norm",
                               "job_id", "position_name",
                               "company", "cosine_score", "rank")

    # Show results in the notebook
    top3.show(truncate=False)

    # Save to MongoDB
    results_pd = top3.toPandas()
    client = MongoClient(MONGO_URI)
    db_stream = client['ats_project']
    docs = results_pd.to_dict(orient='records')
    if docs:
        db_stream['stream_results'].insert_many(docs)
    client.close()

    print(f"Batch {batch_id}: saved {len(docs)} match results to MongoDB")

print("Batch processing function defined.")
```

```python
import time
import pandas as pd
from pyspark.sql import functions as F

def run_pyspark_benchmark(resumes_df, jobs_df, pipeline_model, cosine_udf_func, relevant_categories_list, scale_label):
    timings = {"scale": scale_label}

    # Stage 1: Load + clean
    t0 = time.time()
    r = resumes_df.withColumn(
            "model_text",
            F.concat_ws(" ", F.col("clean_text"),
                        F.coalesce(F.col("category_norm"), F.lit("")))
        )
    j = jobs_df.withColumn(
            "model_text",
            F.concat_ws(" ",
                F.coalesce(F.col("position_name"), F.lit("")),
                F.col("clean_text"),
                F.coalesce(F.col("job_family"), F.lit("")))
        )
    r.count()   # force Spark to actually execute
    j.count()
    timings["load_seconds"] = round(time.time() - t0, 2)

    # Stage 2: TF-IDF vectorization
    t0 = time.time()
    r_tagged = r.withColumn("doc_id",   F.col("resume_id").cast("string")) \
                .withColumn("doc_type", F.lit("resume"))
    j_tagged = j.withColumn("doc_id",   F.col("job_id").cast("string")) \
                .withColumn("doc_type", F.lit("job"))
    corpus_b = r_tagged.select("doc_id","doc_type","model_text") \
               .union(j_tagged.select("doc_id","doc_type","model_text"))
    model_b  = pipeline_model.fit(corpus_b)
    vec_b    = model_b.transform(corpus_b)
    vec_b.count()
    timings["tfidf_seconds"] = round(time.time() - t0, 2)

    #  Stage 3: Similarity matching
    t0 = time.time()
    r_vec = vec_b.filter(F.col("doc_type") == "resume") \
                 .select(F.col("doc_id").alias("resume_id"),
                         F.col("features").alias("resume_features"))
    j_vec = vec_b.filter(F.col("doc_type") == "job") \
                 .select(F.col("doc_id").alias("job_id"),
                         F.col("features").alias("job_features"))

    relevant = r.filter(F.col("category_norm").isin(relevant_categories_list)) \
                .join(r_vec, r["resume_id"].cast("string") == r_vec["resume_id"]) \
                .drop(r_vec["resume_id"])
    j_full   = j.join(j_vec, on="job_id")

    # Sample 50 resumes for timing — enough to measure, fast enough to run
    r_sample = relevant.limit(50) # Use relevant resumes for sampling

    scored = r_sample.crossJoin(j_full) \
                     .withColumn("cosine_score",
                                 cosine_udf_func(F.col("resume_features"),
                                            F.col("job_features"))) \
                     .filter(F.col("cosine_score") > 0.0)
    scored.count()
    timings["match_seconds"] = round(time.time() - t0, 2)

    print(f"Scale {scale_label} done — "
          f"load: {timings['load_seconds']}s | "
          f"tfidf: {timings['tfidf_seconds']}s | "
          f"match: {timings['match_seconds']}s")
    return timings

print("Benchmark function defined.")
```

## Pure Python Benchmarking (Concept Test B)

```python
import time
import json
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity as sk_cosine
import os

def load_json_lines(path):
    with open(path, 'r') as f:
        return [json.loads(line) for line in f if line.strip()]

base_resumes = load_json_lines('data/json/resumes.json')
base_jobs    = load_json_lines('data/json/jobs.json')

print(f"Base data loaded: {len(base_resumes)} resumes, "
      f"{len(base_jobs)} jobs")

def run_benchmark(resumes, jobs, scale_label):
    timings = {"scale": scale_label}

    t0 = time.time()
    resume_texts = [r.get('clean_text', '') or '' for r in resumes]
    job_texts    = [j.get('clean_text', '') or '' for j in jobs]
    resume_texts = [t.lower().strip() for t in resume_texts]
    job_texts    = [t.lower().strip() for t in job_texts]
    timings["load_seconds"] = round(time.time() - t0, 3)

    t0 = time.time()
    vectorizer   = TfidfVectorizer(
        max_features=5000, min_df=2, stop_words='english')
    all_texts    = resume_texts + job_texts
    tfidf_matrix = vectorizer.fit_transform(all_texts)
    resume_vecs  = tfidf_matrix[:len(resume_texts)]
    job_vecs     = tfidf_matrix[len(resume_texts):]
    timings["tfidf_seconds"] = round(time.time() - t0, 3)

    t0 = time.time()
    sample_size  = min(100, len(resume_texts))
    scores       = sk_cosine(resume_vecs[:sample_size], job_vecs)
    top5_indices = np.argsort(scores, axis=1)[:, -5:][:, ::-1]
    pair_count   = sample_size * len(job_texts)
    timings["match_seconds"]  = round(time.time() - t0, 3)
    timings["pairs_scored"]   = pair_count

    print(f"  [{scale_label}] "
          f"load: {timings['load_seconds']}s | "
          f"tfidf: {timings['tfidf_seconds']}s | "
          f"match: {timings['match_seconds']}s | "
          f"pairs: {pair_count:,}")
    return timings

def scale_data(records, id_field, factor):
    scaled = []
    for i in range(factor):
        for rec in records:
            copy = dict(rec)
            copy[id_field] = str(rec.get(id_field, '')) + f'_s{i}'
            scaled.append(copy)
    return scaled

benchmark_results_python = []
print("\nRunning Python benchmarks...\n")

for factor in [1, 5, 10]:
    r = scale_data(base_resumes, 'resume_id', factor)
    j = scale_data(base_jobs,    'job_id',    factor)
    result = run_benchmark(r, j, f"{factor}x")
    benchmark_results_python.append(result)

# ← renamed to bench_df_python so it never overwrites bench_df_spark
bench_df_python = pd.DataFrame(benchmark_results_python)

os.makedirs('outputs/charts', exist_ok=True)
bench_df_python.to_csv(
    'outputs/charts/benchmark_summary_python.csv', index=False)

print("\n=== Python Benchmark Summary ===")
print(bench_df_python[["scale", "load_seconds",
                        "tfidf_seconds", "match_seconds",
                        "pairs_scored"]].to_string(index=False))
```
